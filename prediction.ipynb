```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
import pickle
import pandas as pd

# Saved encoder ko load karo
with open('onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encoder_geo = pickle.load(file)
```

```python
### load the trained model,scaler pickle,one hot
model=load_model('model.h5')

## load the encoder and scaler
with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)

with open('label_encode_gender.pkl','rb') as file:
    label_encode_gender=pickle.load(file)


with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
## example input data
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000

}
```

```python
import pandas as pd

# Dictionary ko dataframe mein badlo
input_df = pd.DataFrame([input_data])
```

```python
# Geography column ko transform karo
geo_encoded = onehot_encoder_geo.transform(input_df[['Geography']]).toarray()

# Isko ek naye dataframe mein badlo jiske column names wahi hon jo training mein the
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
# 1. Purana 'Geography' hatao aur naye columns jodo
input_df = pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)

# 2. Ab check karo, ab France gayab ho jayega aur 0/1 dikhega
input_df
```

```python
input_df
```

```python
input_df['Gender']=label_encode_gender.transform(input_df['Gender'])
input_df
```

```python
## scaling the data
input_scaled=scaler.transform(input_df)
```

```python
input_scaled
```

```python
## prediction
prediction=model.predict(input_scaled)
prediction
```
